```python
import cv2
from matplotlib import pyplot as plt

# Inisialisasi classifier untuk deteksi wajah
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Baca gambar JPG
image = cv2.imread('rifki5.jpg')

# Konversi gambar ke skala abu-abu (grayscale) untuk deteksi wajah
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Lakukan deteksi wajah
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Buat salinan gambar untuk menampilkan kotak hijau hanya pada gambar hasil akhir
image_with_faces = image.copy()

# Gambar kotak di sekitar wajah yang terdeteksi pada gambar hasil akhir
for (x, y, w, h) in faces:
    cv2.rectangle(image_with_faces, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Menampilkan gambar sebelum deteksi wajah
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Before')
plt.axis('off')

# Menampilkan gambar setelah deteksi wajah
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_with_faces, cv2.COLOR_BGR2RGB))
plt.title('After')
plt.axis('off')

plt.show()
```
